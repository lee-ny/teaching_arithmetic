```python
import numpy as np
import pandas as pd

import math
```

```python
with open('add_examples_10000.txt', 'r') as f:
    lines = f.readlines()
```

```python
x_list = []
y_list = []
for line in lines:
    x, y = line.split('+')
    x_list.append(int(x))
    y = y.split('=')[0]
    y_list.append(int(y))
```

```python
x_set = set(x_list)
y_set = set(y_list)
total_set = set(x_list + y_list)
```

```python
print(x_set)
```

```python
len(y_set)
```

```python
count_x_dict = {}
count_y_dict = {}

min_x_count = 1000
min_y_count = 1000
min_x = -1
min_y = -1

for x in x_set:
    count_x = len([i for i in x_list if i == x])
    count_y = len([i for i in y_list if i == x])
    count_x_dict[x] = count_x
    count_y_dict[x] = count_y
    
    if count_x < min_x_count:
        min_x_count = count_x
        min_x = x
    if count_y < min_y_count:
        min_y_count = count_y
        min_y = x
```

```python
print(min_x, min_y)
print(min_x_count, min_y_count)
```

```python
print(count_x_dict)
print(count_y_dict)
```

```python
x_count_sum = sum(count_x_dict.values())
y_count_sum = sum(count_y_dict.values())

print(x_count_sum, y_count_sum)
```

```python
x_count_df = pd.DataFrame(count_x_dict)
y_count_df = pd.DataFrame(count_y_dict)
```

```python
my_dict = {'x':[], 'count_x':[], 'count_y':[], 'count_x_and_y' : []}
for key in count_x_dict.keys():
    my_dict['x'].append(key)
    my_dict['count_x'].append(count_x_dict[key])
    my_dict['count_y'].append(count_y_dict[key])
    my_dict['count_x_and_y'].append(count_x_dict[key] + count_y_dict[key])
```

```python
my_df = pd.DataFrame(my_dict)
```
